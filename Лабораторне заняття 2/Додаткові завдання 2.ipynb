## 1 Візуалізація емпіричної імовірності
Побудуйте графік залежності емпіричної імовірності випадіння герба (HEAD) від кількості підкидань чесної монети (симуляцій).

**Зауваження.** Завдання передбачає використання бібліотеки [Matplotlib](https://matplotlib.org/).

```python
# Місце для Вашого коду
def getProbably(n, w=["H","T"],ev="H"):
    tosses = [w[i] for i in np.random.randint(len(w), size=n)]
    return tosses.count(ev)/len(tosses)
```

```python
count_tosses = np.arange(1000, 100000, 1000)
prob = np.array([getProbably(n) for n in count_tosses])
fig, ax = plt.subplots()
ax.plot(count_tosses, prob)

ax.set(xlabel='count of tosses (n) ', ylabel='probability of HEAD',
       title='графік залежності емпіричної імовірності випадіння герба (HEAD) від кількості підкидань')
ax.grid()

fig.savefig("probability_HEAD.png")
plt.show()
```
